# edge-detect fixed Kernel

```python
# load in dataset
datapath = '../../../../../Desktop/mnist/'
data = np.loadtxt(datapath + 'mnist_mini_train.csv',delimiter = ',')

# MOVE TO BACKEND
# contrast normalize image data
def contrast_normalize(data):
    data_means = np.mean(data,axis = 0)
    data = data - data_means
    data_stds = np.std(data,axis = 0)
    data = data/data_stds
    return data

# normalize the input 
x = data[:,:-1]
y = data[:,-1:]
x_decontrast = contrast_normalize(x.T).T
```

```python
# make square version of data 
# MOVE TO BACKEND
X_square = np.zeros((len(y),28,28))
for i in range(0,len(y)):
    X_square[i,:,:] = np.reshape(x_decontrast[i,:],(28,28),1)
```

```python
# load kernels
kernels = convlib.image_viz.load_kernels()

####### swtiches for frontend API #########
# params
sliding_window_size= (6,6) 
stride=3
pooling_func= 'max'
#normalization = 'l2'
####### switches for frontend API ########

# get number of images in the dataset
# MOVE TO BACKEND
num_images = np.shape(X_square)[0]
        
# a test run to find the number of features with the params above
# MOVE TO BACKEND
test = convlib.image_viz.make_feat(X_square[0,:,:], kernels, sliding_window_size=sliding_window_size, stride=stride)
num_features = np.shape(test)[0]       
```

```python
# produce features for input data
# MOVE TO BACKEND
feat = np.zeros((num_features,num_images))
for i in range(0,num_images):
    # extract features
    feat[:,i] = convlib.image_viz.make_feat(X_square[i,:,:], kernels, sliding_window_size=sliding_window_size,
                                            stride=stride, pooling_func=pooling_func)
```

```python
data_new = np.concatenate((feat.T,y),axis = 1)
```

```python
data_train = data_new[:4000,:]
data_test = data_new[4000:,:]
```

```python
demo.input_data(data_train,data_test,normalize = True)

# choose cost function
demo.choose_cost(cost_name = 'multiclass_softmax')

# setup network architecture
activation_name = 'maxout'
layer_sizes = [np.shape(data_train)[1]-1,10,10,10]
demo.architecture_settings(activation_name,layer_sizes)

# setup optimizer
demo.optimizer_settings(alpha = 10**(-1),max_its = 500,version = 'normalized',scale = 0.5)

# fit
demo.fit(verbose = True)

# compute cost plots on training and testing data
demo.compute_cost_plots()

# produce cost functio plots for training and testing data
demo.plot_histories(start = 0)
```
