```python
    def make_cos_approx(signal,num_elements):
        f = []
        for n in np.arange(0,num_elements,3):
            if np.mod(n,2) == 0:
                f.append((1/float(n+1)**2)*np.cos(2*np.pi*n*signal))
            else:
                f.append((1/float(n+1)**2))*np.sin(2*np.pi*n*signal))
                
        f = np.asarray(f)
        f.shape = (np.shape(f)[0], np.shape(f)[1])
        f = f.T
        return f
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
r = np.linspace(0,1,300)[:, np.newaxis]
f = make_cos_approx(r,10)
t = np.sum(f,axis = 1)
plt.plot(r,t)
```
